```python
bettings_df_columns_dirty = cursor.fetchall()
```

```python
bettings_df_columns_clean = []
for list in bettings_df_columns_dirty:
    bettings_df_columns_clean.append(list[0])
```

```python
bettings_df_columns_clean
```

```python
betting_df.columns = bettings_df_columns_clean
```

```python
betting_df.head()
```

```python
betting_df = betting_df.loc[:, 'AccountNo' : 'Product']
```

```python
betting_df = betting_df \
    .assign(
        Lost_Amt = betting_df['Bet_Amt'] - betting_df['Win_Amt']
    )
```

```python
betting_num_cols = betting_df.select_dtypes(include='number').columns

betting_num_cols = [col for col in betting_num_cols if col != 'CategoryId']

for col in betting_num_cols:
    plt.figure()
    sns.histplot(betting_df[col])
    plt.title(col)
    plt.show()
```

```python
betting_df.query('Lost_Amt < 0')

# must be won bets
```

```python
product_matrix = betting_df \
    .groupby('Product') \
    .agg({
        'Bet_Amt' : 'sum',
        'Win_Amt' : 'sum',
        'Lost_Amt' : 'sum'
    }) \
    .sort_values('Bet_Amt', ascending=False) \
    .assign(
        Margin = lambda x: x['Lost_Amt'] / x['Bet_Amt']
    ) 
```

```python
product_matrix
```

```python
product_matrix['Bet_Amt'].plot.bar()

plt.title('Bet Amount by Product');
```

```python
product_matrix['Lost_Amt'].plot.bar();

plt.title('Lost Amount');
```

```python
product_matrix[['Bet_Amt', 'Lost_Amt']].plot.bar();

plt.title('Bet Amount VS Lost ');
```

```python
sns.heatmap(
    product_matrix['Margin'].to_frame(),
    vmin = 0,
    vmax=1,
    cmap='coolwarm'
);

plt.title('Margin Heatmap');
```

```python
betting_df['BetDate'] = pd.to_datetime(betting_df['BetDate'])
```

```python
betting_df \
    .groupby('BetDate') \
    .agg({'BetCount' : 'sum'}) \
    .plot(
        title='Count of Bets',
        legend=False);

# betting activity peaks at throughout November & end of december
```

```python
betting_df \
    .groupby('BetDate') \
    .agg({'Lost_Amt' : 'sum'}) \
    .plot(
        title='Lost Amount',
        legend=False);

# negative revenue in mid october 
# low revenueend in december despite high bet volume
# revenue peaks around end of november
```

```python
betting_df \
    .groupby(betting_df['BetDate']) \
    .agg({'Lost_Amt' : 'sum'}) \
    .sort_values('Lost_Amt') \
    .head(5)

# 3 of top 5 negative profit days occured in October
```
